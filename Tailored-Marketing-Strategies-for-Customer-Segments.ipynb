```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veri setini yükleme
df = pd.read_csv("data_synthetic.csv")

# Segmentasyon gruplarını demografik ve finansal özelliklere göre analiz etme
segment_summary = df.groupby('Segmentation Group').agg(
    Avg_Age=('Age', 'mean'),
    Avg_Income_Level=('Income Level', 'mean'),
    Avg_Credit_Score=('Credit Score', 'mean'),
    Avg_Premium_Amount=('Premium Amount', 'mean'),
    Avg_Deductible=('Deductible', 'mean'),
    Total_Claims=('Claim History', 'sum')
).reset_index()

# Cinsiyet dağılımı
gender_by_segment = pd.crosstab(df['Segmentation Group'], df['Gender'], normalize='index') * 100
marital_status_by_segment = pd.crosstab(df['Segmentation Group'], df['Marital Status'], normalize='index') * 100

# Sürücü Kaydı dağılımı
driving_record_by_segment = pd.crosstab(df['Segmentation Group'], df['Driving Record'], normalize='index') * 100
print("\nSürücü Kaydına Göre Dağılımı (%):")
print(driving_record_by_segment)

# Risk Profili dağılımı
risk_profile_by_segment = pd.crosstab(df['Segmentation Group'], df['Risk Profile'], normalize='index') * 100
print("\nRisk Profiline Göre Dağılımı (%):")
print(risk_profile_by_segment)

# En sık tercih edilen iletişim kanalı
preferred_communication_by_segment = df.groupby('Segmentation Group')['Preferred Communication Channel'].agg(lambda x: x.mode()[0])
print("\nEn Sık Kullanılan İletişim Kanalı:")
print(preferred_communication_by_segment)

# Ortalama yaş
plt.figure(figsize=(10, 6))
sns.barplot(x='Segmentation Group', y='Avg_Age', data=segment_summary, hue='Segmentation Group', palette='viridis', legend=False)
plt.title('Segmentasyon Gruplarına Göre Ortalama Yaş')
plt.xlabel('Segmentasyon Grubu')
plt.ylabel('Ortalama Yaş')
plt.show()

# Ortalama kredi skoru
plt.figure(figsize=(10, 6))
sns.barplot(x='Segmentation Group', y='Avg_Credit_Score', data=segment_summary, hue='Segmentation Group', palette='magma', legend=False)
plt.title('Segmentasyon Gruplarına Göre Ortalama Kredi Skoru')
plt.xlabel('Segmentasyon Grubu')
plt.ylabel('Ortalama Kredi Skoru')
plt.show()

# Ortalama prim miktarı
plt.figure(figsize=(10, 6))
sns.barplot(x='Segmentation Group', y='Avg_Premium_Amount', data=segment_summary, hue='Segmentation Group', palette='cividis', legend=False)
plt.title('Segmentasyon Gruplarına Göre Ortalama Prim Miktarı')
plt.xlabel('Segmentasyon Grubu')
plt.ylabel('Ortalama Prim Miktarı')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Veri setini yükleme
df = pd.read_csv("data_synthetic.csv")

# Kümeleme için kullanılacak özellikleri seçme
features = ['Age', 'Income Level', 'Credit Score', 'Premium Amount', 'Claim History']
X = df[features]

# Veriyi standartlaştırma
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# En uygun küme sayısını (k) bulmak için "dirsek (elbow)" yöntemini kullanma
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

# Dirsek eğrisini çizme
plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Küme Sayısı (k)')
plt.ylabel('Hata Kareleri Toplamı (SSE)')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Veri setini yükleme
df = pd.read_csv("data_synthetic.csv")

# Kümeleme için kullanılacak özellikleri seçme
features = ['Age', 'Income Level', 'Credit Score', 'Premium Amount', 'Claim History']
X = df[features]

# Veriyi standardize etme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# En uygun küme sayısını (k=4) kullanarak K-Means modelini oluşturma ve eğitme
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')
df['New_Segment'] = kmeans.fit_predict(X_scaled)

# Yeni segmentlerin özelliklerini analiz etme
new_segment_summary = df.groupby('New_Segment').agg(
    Avg_Age=('Age', 'mean'),
    Avg_Income_Level=('Income Level', 'mean'),
    Avg_Credit_Score=('Credit Score', 'mean'),
    Avg_Premium_Amount=('Premium Amount', 'mean'),
    Avg_Claim_History=('Claim History', 'mean')
).reset_index()

print("Yeni Oluşturulan Segmentlerin Özellikleri:")
print(new_segment_summary)

# Yeni segmentlerin sürüş kaydı dağılımını inceleme
driving_record_by_new_segment = pd.crosstab(df['New_Segment'], df['Driving Record'], normalize='index') * 100
print("\nYeni Segmentlere Göre Sürüş Kaydı Dağılımı (%):")
print(driving_record_by_new_segment)

# Yeni segmentlerin yaşam olayları dağılımını inceleme
life_events_by_new_segment = pd.crosstab(df['New_Segment'], df['Life Events'], normalize='index') * 100
print("\nYeni Segmentlere Göre Yaşam Olayları Dağılımı (%):")
print(life_events_by_new_segment)

# Ortalamaları görselleştirme
# Yeni segmentlere göre ortalama yaş
plt.figure(figsize=(10, 6))
sns.barplot(x='New_Segment', y='Avg_Age', data=new_segment_summary, palette='viridis')
plt.title('Yeni Segmentlere Göre Ortalama Yaş')
plt.xlabel('Yeni Segment')
plt.ylabel('Ortalama Yaş')
plt.show()


# Yeni segmentlere göre ortalama prim miktarı
plt.figure(figsize=(10, 6))
sns.barplot(x='New_Segment', y='Avg_Premium_Amount', data=new_segment_summary, palette='magma')
plt.title('Yeni Segmentlere Göre Ortalama Prim Miktarı')
plt.xlabel('Yeni Segment')
plt.ylabel('Ortalama Prim Miktarı')
plt.show()


# Yaşa ve gelir seviyesine göre yeni segmentleri görselleştirme
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Age', y='Income Level', hue='New_Segment', data=df, palette='viridis', style='New_Segment', s=50)
plt.title('Yaş ve Gelir Seviyesine Göre Yeni Müşteri Segmentleri')
plt.xlabel('Yaş')
plt.ylabel('Gelir Seviyesi')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Önceki adımdan oluşturulan veri setini yükleme.
# Bu veri seti, yeni segment etiketlerini içermektedir.
df = pd.read_csv("customer_segments_new.csv")

# Yeni hedef değişkeni 'Has_Claim'i oluşturma.
# Eğer 'Claim History' 0'dan büyükse 1, aksi halde 0 olacak.
df['Has_Claim'] = df['Claim History'].apply(lambda x: 1 if x > 0 else 0)

# Özellikleri (X) ve hedef değişkeni (y) belirleme.
# Yeni oluşturduğumuz segment etiketini de modelin bir özelliği olarak kullanıyoruz.
features = ['Age', 'Income Level', 'Credit Score', 'Premium Amount', 'Risk Profile', 'New_Segment']
X = df[features]
y = df['Has_Claim']

# Veri setini eğitim ve test olarak ayırma.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Eğitim Seti Boyutu: {X_train.shape[0]} satır, {X_train.shape[1]} sütun")
print(f"Test Seti Boyutu: {X_test.shape[0]} satır, {X_test.shape[1]} sütun")

# Rastgele Orman (Random Forest) Sınıflandırıcısı modelini eğitme.
model_claim = RandomForestClassifier(n_estimators=100, random_state=42)
print("\nHasar Tahmin Modeli Eğitimi Başlıyor...")
model_claim.fit(X_train, y_train)
print("Model Eğitimi Tamamlandı.")

# Modelin performansını test seti üzerinde değerlendirme.
y_pred_claim = model_claim.predict(X_test)

# Karışıklık matrisini hesaplama ve görselleştirme.
cm_claim = confusion_matrix(y_test, y_pred_claim)
print("\nHasar Tahmin Modeli Karışıklık Matrisi:")
print(cm_claim)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_claim, annot=True, fmt='d', cmap='Blues', xticklabels=['Hasar Yok', 'Hasar Var'], yticklabels=['Hasar Yok', 'Hasar Var'])
plt.title('Hasar Tahmin Modeli Karışıklık Matrisi')
plt.ylabel('Gerçek Değerler')
plt.xlabel('Tahmin Edilen Değerler')
plt.show()


# Sınıflandırma raporunu yazdırma.
print("\nHasar Tahmin Modeli Sınıflandırma Raporu:")
print(classification_report(y_test, y_pred_claim))

# Özellik önemini analiz etme.
feature_importances_claim = pd.DataFrame(model_claim.feature_importances_, index=X_train.columns, columns=['importance']).sort_values('importance', ascending=False)
print("\nHasar Tahmininde En Önemli Özellikler:")
print(feature_importances_claim)
```

```python
# Gerekli kütüphaneleri içe aktarma
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

# Önceki analizden kaydedilen veri setini yükleme
df = pd.read_csv("customer_segments_new.csv")

# Yeni hedef değişkeni oluşturma: Hasar varsa 1, yoksa 0
df['Has_Claim'] = df['Claim History'].apply(lambda x: 1 if x > 0 else 0)

# Özellikler (X) ve hedef değişken (y) olarak veriyi ayırma
# Yeni segment sütununu ve ilgili diğer özellikleri kullanma
features = ['Age', 'Income Level', 'Credit Score', 'Premium Amount', 'Deductible', 'Risk Profile', 'New_Segment', 'Claim History']
X = df[features]
y = df['Has_Claim']

# Kategorik özellikleri one-hot encoding ile dönüştürme (gerekli ise)
# Bu veri setinde 'New_Segment' ve diğer özellikler zaten sayısal.

# Veriyi eğitim ve test setlerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print("--- Hasar Tahmini Modeli Oluşturuluyor ---")
# RandomForestClassifier modelini oluşturma
model = RandomForestClassifier(n_estimators=100, random_state=42)

print("Model Eğitimi Başlıyor...")
model.fit(X_train, y_train)
print("Model Eğitimi Tamamlandı.")

# Test seti üzerinde tahmin yapma
y_pred = model.predict(X_test)

# Model performansını değerlendirme
cm = confusion_matrix(y_test, y_pred)
print("\nKarışıklık Matrisi:")
print(cm)

print("\nSınıflandırma Raporu:")
print(classification_report(y_test, y_pred))

# Karışıklık matrisini görselleştirme
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Hasar Yok', 'Hasar Var'], yticklabels=['Hasar Yok', 'Hasar Var'])
plt.title('Hasar Tahmini Modeli Karışıklık Matrisi')
plt.ylabel('Gerçek Değerler')
plt.xlabel('Tahmin Edilen Değerler')
plt.show()
```

Bu proje, müşteri verilerini analiz ederek K-Means ile 4 yeni müşteri segmenti oluşturmuştur. Ardından, bu segmentler ve diğer müşteri bilgileri kullanılarak, bir müşterinin hasar talebinde bulunma olasılığını tahmin eden bir makine öğrenimi modeli geliştirilmiştir. Proje, bu analizlere dayanarak her bir segmente özel pazarlama stratejileri sunmaktadır.
